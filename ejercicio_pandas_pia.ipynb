# **Preparing dataset**

---



```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv("Telco.csv")
```

***1.- Convert the column ‘TotalCharges’ to a numeric format, coercing errors if necessary***

```python
data.info()
```

```python
data['TotalCharges'] = pd.to_numeric(data['TotalCharges'],errors='coerce')
```

```python
data.info()
```

***2.- Change all column names to lowercase.***

```python
data.columns = data.columns.str.lower()
data.columns
```

***3.- For string values change spaces to an underscore symbol***

```python
data = data.replace(" ", '_', regex=True)
data
```

***4.- Convert the column ‘churn’ to integer format (no - 0, yes - 1).***

```python
data = data.replace("No", '0', regex=True)
data
data = data.replace("Yes", '1', regex=True)
data
data['churn'] = pd.to_numeric(data['churn'],errors='coerce')
```

```python
data['churn']
```

***5.-Check for categorical columns how many different values they have.***

```python
categoricas = [
    'gender','multiplelines','internetservice','onlinesecurity',
    'onlinebackup','deviceprotection','techsupport','streamingtv',
    'streamingmovies','contract','paymentmethod'
    ]
for i in categoricas:
    print(data[i].value_counts())
```

***6.- Delete the column ‘customerid’ from the dataset.***

```python
data.drop(['customerid'], axis=1)
```
